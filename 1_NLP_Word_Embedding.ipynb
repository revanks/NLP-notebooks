#Word Embedding

```
import numpy as np
```

#Quote by Richard Feynman

```
Quote=['We are at the very beginning of time for the human race',
        'It is not unreasonable that we grapple with problems', 
        'But there are tens of thousands of years in the future', 
        'Our responsibility is to do what we can, learn what we can', 
        'improve the solutions, and pass them on',
        'Richard P. Feynman']
Quote
```

#Choose vocab size

```
### Vocabulary size
voc_size=10000
```

#One hot encoding all words

```
from tensorflow.keras.preprocessing.text import one_hot
onehot_repr=[one_hot(words,voc_size)for words in Quote] 
print(onehot_repr)
```

#Pre padding sentences

```
from tensorflow.keras.preprocessing.sequence import pad_sequences
sent_length=14
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

#Word Embedding

```
from tensorflow.keras.layers import Embedding
from tensorflow.keras.models import Sequential
dim=10 #NO of Features
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile('adam','mse')
model.summary()
```

```
print(model.predict(embedded_docs))
```
